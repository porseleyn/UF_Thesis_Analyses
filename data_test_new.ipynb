```python
# Filter only Time_on_offense_teamA state rows
a_offense_rows = df[df['Behaviour'] == 'Time_on_offense_teamA']

# Get list of start/stop timestamps in order
starts = a_offense_rows[a_offense_rows['B_Type'] == 'START']['Time'].tolist()
stops  = a_offense_rows[a_offense_rows['B_Type'] == 'STOP']['Time'].tolist()

# Safety check: they should match in length
if len(starts) != len(stops):
    raise ValueError("Mismatched number of START and STOP events for Team A offense.")

# Zip into intervals
a_intervals = list(zip(starts, stops))
```

```python
# Filter only Time_on_offense_teamB state rows
b_offense_rows = df[df['Behaviour'] == 'Time_on_offense_teamB']

# Get list of start/stop timestamps in order
starts = b_offense_rows[b_offense_rows['B_Type'] == 'START']['Time'].tolist()
stops  = b_offense_rows[b_offense_rows['B_Type'] == 'STOP']['Time'].tolist()

# Safety check: they should match in length
if len(starts) != len(stops):
    raise ValueError("Mismatched number of START and STOP events for Team B offense.")

# Zip into intervals
b_intervals = list(zip(starts, stops))
```

```python
def subtract_intervals(base_intervals, sub_intervals):
    result = []
    for start, stop in base_intervals:
        temp = [(start, stop)]
        for sub_start, sub_stop in sub_intervals:
            new_temp = []
            for s, e in temp:
                # Skip non-overlapping
                if sub_stop <= s or sub_start >= e:
                    new_temp.append((s, e))
                else:
                    # Chop out the timeout
                    if sub_start > s:
                        new_temp.append((s, sub_start))
                    if sub_stop < e:
                        new_temp.append((sub_stop, e))
            temp = new_temp
        result.extend(temp)
    return result

# Get timeout intervals
timeout_rows = df[df['Behaviour'] == 'Timeout']
timeout_starts = timeout_rows[timeout_rows['B_Type'] == 'START']['Time'].tolist()
timeout_stops  = timeout_rows[timeout_rows['B_Type'] == 'STOP']['Time'].tolist()
timeout_intervals = list(zip(timeout_starts, timeout_stops))

# Subtract timeouts from Team A offense
a_intervals_clean = subtract_intervals(a_intervals, timeout_intervals)
b_intervals_clean = subtract_intervals(b_intervals, timeout_intervals)

# Sum offense duration team A
a_total_seconds = sum(stop - start for start, stop in a_intervals_clean)
a_total_minutes = a_total_seconds/60

print("Team A was on offense for:", a_total_minutes, "minutes (excluding timeouts).")

# Sum offense duration team B
b_total_seconds = sum(stop - start for start, stop in b_intervals_clean)
b_total_minutes = b_total_seconds/60

print("Team B was on offense for:", b_total_minutes, "minutes (excluding timeouts).")
```

### Indicator 9: Possession efficiency.
Number of passes in each successful scoring attempt.

```python
# Step 1: Extract Team A possession intervals (START–STOP)
offense = df[df['Behaviour'] == 'Time_on_offense_teamA']
starts = offense[offense['Behaviour type'] == 'START']['Time'].tolist()
stops = offense[offense['Behaviour type'] == 'STOP']['Time'].tolist()

starts.sort()
stops.sort()

if len(starts) != len(stops):
    raise ValueError("Mismatched START/STOP counts for Team A offense.")

teamA_possessions = list(zip(starts, stops))

# Step 2: Create Team B possessions as gaps between Team A's possessions
teamB_possessions = []
for i in range(min(len(stops), len(starts) - 1)):
    b_start = stops[i]
    b_stop = starts[i + 1]
    if b_stop > b_start:
        teamB_possessions.append((b_start, b_stop))

# Step 3: Get timestamps of all scoring events
score_times = df[df['Behaviour'] == 'Score']['Time'].tolist()

# Step 4: Identify possessions that included a score
def get_scoring_possessions(possessions, score_events):
    scoring_possessions = []
    for start, stop in possessions:
        if any((score_time >= start) and (score_time <= stop) for score_time in score_events):
            scoring_possessions.append((start, stop))
    return scoring_possessions

scoring_pos_A = get_scoring_possessions(teamA_possessions, score_times)
scoring_pos_B = get_scoring_possessions(teamB_possessions, score_times)

# Step 5: Count passes in each scoring possession
def count_passes_in_possessions(possessions, team_df):
    pass_counts = []
    for start, stop in possessions:
        passes = team_df[
            (team_df['Time'] >= start) &
            (team_df['Time'] <= stop) &
            (team_df['Behaviour'] == 'Successful_pass')
        ]
        pass_counts.append(len(passes))
    return pass_counts

# Step 6: Get pass counts per scoring possession
passes_per_score_A = count_passes_in_possessions(scoring_pos_A, df_teamA_offense)
passes_per_score_B = count_passes_in_possessions(scoring_pos_B, df_teamB_offense)

# Step 7: Compute averages
avg_passes_per_score_A = np.mean(passes_per_score_A) if passes_per_score_A else 0
avg_passes_per_score_B = np.mean(passes_per_score_B) if passes_per_score_B else 0

# Step 8: Display results
print(f"Team A: Avg. passes per scoring possession = {avg_passes_per_score_A:.2f}")
print(f"Team B: Avg. passes per scoring possession = {avg_passes_per_score_B:.2f}")
```

### Indicator 12: Possession duration.
Time in seconds a team holds possession before scoring or losing the disc.

So how much time in minutes each offense was for each team, and then the average of that.

```python
# Step 1: Extract Team A possession intervals (START–STOP)
a_offense = df[df['Behaviour'] == 'Time_on_offense_teamA']
a_starts = a_offense[a_offense['B_Type'] == 'START']['Time'].tolist()
a_stops = a_offense[a_offense['B_Type'] == 'STOP']['Time'].tolist()

a_starts.sort()
a_stops.sort()

if len(a_starts) != len(a_stops):
    raise ValueError("Mismatched START/STOP counts for Team A offense.")

teamA_possessions = list(zip(a_starts, a_stops))

# Step 2: Extract Team A possession intervals (START–STOP)
b_offense = df[df['Behaviour'] == 'Time_on_offense_teamB']
b_starts = b_offense[b_offense['B_Type'] == 'START']['Time'].tolist()
b_stops = b_offense[b_offense['B_Type'] == 'STOP']['Time'].tolist()

b_starts.sort()
b_stops.sort()

if len(b_starts) != len(b_stops):
    raise ValueError("Mismatched START/STOP counts for Team B offense.")

teamB_possessions = list(zip(b_starts, b_stops))

# Step 3: Extract timeout intervals
timeouts = df[df['Behaviour'] == 'Timeout']
to_starts = timeouts[timeouts['B_Type'] == 'START']['Time'].tolist()
to_stops  = timeouts[timeouts['B_Type'] == 'STOP']['Time'].tolist()

if len(to_starts) != len(to_stops):
    raise ValueError("Mismatched START/STOP counts for timeouts.")

timeout_intervals = list(zip(to_starts, to_stops))

# Step 4: Function to subtract timeout overlaps from possession intervals
def subtract_timeouts(possessions, timeouts):
    clean_durations = []
    for start, stop in possessions:
        duration = stop - start
        for t_start, t_stop in timeouts:
            overlap_start = max(start, t_start)
            overlap_end = min(stop, t_stop)
            if overlap_start < overlap_end:
                duration -= (overlap_end - overlap_start)
        clean_durations.append(duration)
    return clean_durations

# Step 5: Subtract timeouts from each team's possessions
teamA_durations = subtract_timeouts(teamA_possessions, timeout_intervals)
teamB_durations = subtract_timeouts(teamB_possessions, timeout_intervals)

# Step 6: Calculate averages
avg_teamA_sec = np.mean(teamA_durations)
avg_teamB_sec = np.mean(teamB_durations)

# Step 7: Display results
print(f"Team A average possession duration: {avg_teamA_sec:.2f} seconds ({avg_teamA_sec / 60:.2f} minutes)")
print(f"Team B average possession duration: {avg_teamB_sec:.2f} seconds ({avg_teamB_sec / 60:.2f} minutes)")
```
